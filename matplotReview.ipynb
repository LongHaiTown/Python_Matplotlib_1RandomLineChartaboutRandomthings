<h1> RANDOM LINE CHART about RANDOM THINGS in years

## IMPORT 

```python
from matplotlib import pyplot as plt
import random
```

Create a data (RANDOM LINE CHART about RANDOM THINGS in years)

```python
years = range(1999,2011)
thing1 = []
for i in range(0,12):
    thing1.append(random.randrange(400,700))
thing2 =[]
for i in range(0,12):
    thing2.append(random.randrange(400,700))
```

# PLOT THE DATA

```python
plt.plot(years,thing1,marker= 'o')
plt.plot(years,thing2,color = 'pink', marker= 'o')
plt.xlabel("YEARs")
plt.ylabel("Random statistics")
plt.title("RANDOM LINE CHART about RANDOM THINGS in years")
plt.legend(["Number of thing 1","Number of thing 2"],loc =1)
plt.show()
```
